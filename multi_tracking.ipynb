```python
import cv2
import matplotlib.pyplot as plt
from random import randint
#file names 
file1 = "motocross.mp4"
file2 = "dogs_playing.mp4"

#file used
file = file2
output_file = "tracking_video_" + file
print(output_file)

bblist = []
color = []    #color for bounding boxes

#Tracker names
tracker = ['BOOSTING', 'MIL', 'KCF','TLD', 'MEDIANFLOW', 'GOTURN', 'MOSSE', 'CSRT']
tracker_name = 'MIL'

#create multi-tracker object
Multi_tracker = cv2.MultiTracker_create()


def create_tracker(tracker_name):
    tracker = ['BOOSTING', 'MIL', 'KCF','TLD', 'MEDIANFLOW', 'GOTURN', 'MOSSE', 'CSRT']
    if tracker_name == tracker[0]:
        tracker = cv2.TrackerBoosting_create()
    elif tracker_name == tracker[1]:
        tracker = cv2.TrackerMIL_create()
    elif tracker_name == tracker[2]:
        tracker = cv2.TrackerKCF_create()
    elif tracker_name == tracker[3]:
        tracker = cv2.TrackerTLD_create()
    elif tracker_name == tracker[4]:
        tracker = cv2.TrackerMedianFlow_create()
    elif tracker_name == tracker[5]:
        tracker = cv2.TrackerGOTURN_create()
    elif tracker_name == tracker[6]:
        tracker = cv2.TrackerMOSSE_create()
    elif tracker_name == tracker[7]:
        tracker = cv2.TrackerCSRT_create()
    else:
        tracker = None
    return tracker

print("Tracker being used is:",tracker_name)

cap = cv2.VideoCapture(file)

fps = 30
video_size = (960,720)
# Create a VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
save_output = cv2.VideoWriter(output_file,  fourcc, fps, video_size) 

# Capture the first video frame
ret, frame = cap.read() 

if ret:
    #select all bounding boxes and press "s" to save locations
    while(True):
        bbox = cv2.selectROI("MULTI-TRACKER",frame,False)  #select multiple ROIs
        bblist.append(bbox)    #save locations
        color.append((randint(64, 255), randint(64, 255), randint(64, 255)))
        print("Press enter to complete  twice to save the bounding box")
        print("Press 'esc' to exit. ")
#         if cv2.waitKey() & 0xff == ord('s'):
        if cv2.waitKey(0) == 27:
            break
    cv2.destroyAllWindows()

    #add to multi-tracker object
for b in bblist:
    Multi_tracker.add(create_tracker(tracker_name),frame,b)

# print(cap.isOpened())
while cap.isOpened():
    ret,frame = cap.read()
    if not ret:
        break
    state, boxes = Multi_tracker.update(frame)
    
    for i,box in enumerate(boxes):
        #get box x,y,w,h
        p1 = (int(box[0]), int(box[1]))
        p2 = (int(box[0] + box[2]), int(box[1] + box[3]))
        cv2.rectangle(frame, p1, p2, color[i], 3)  #create rectangle
        
    save_output.write(frame)   #save video
    cv2.imshow("frame",frame)

    if cv2.waitKey(1) == 27:
        break
print("done")
#release video objects
cap.release()
save_output.release()
cv2.destroyAllWindows()
```
